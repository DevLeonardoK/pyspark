```python
from sqlalchemy.orm import declarative_base, sessionmaker
from sqlalchemy import Column, INTEGER, VARCHAR,create_engine
```

```python
from config import env_vars
```

```python
try:
    
    engine = create_engine(f"postgresql+psycopg2://{env_vars.database_username.get_secret_value()}:{env_vars.database_password.get_secret_value()}@{env_vars.database_url.get_secret_value()}/{env_vars.database_name.get_secret_value()}")
    
    engine.connect()
    
    Base = declarative_base()
    
    db_session = sessionmaker(engine)
    
except Exception as e:
    raise e
```

```python
class JobSalary(Base):
    __tablename__ = "job_salary"
    id = Column(INTEGER, primary_key=True, unique=True, nullable=False, autoincrement=True)
    job_title = Column(VARCHAR, nullable=False)
    experience_years = Column(INTEGER, nullable=False)
    education_level = Column(VARCHAR, nullable=False)
    skills_count = Column(INTEGER, nullable=False)
    industry = Column(VARCHAR, nullable=False)
    company_size = Column(VARCHAR, nullable=False)
    location = Column(VARCHAR, nullable=False)
    remote_work = Column(VARCHAR, nullable=False)
    certifications = Column(INTEGER, nullable=False)
    salary = Column(INTEGER, nullable=False)
    
Base.metadata.create_all(engine)
```
