```python
from pydantic import BaseModel, Field
```

```python
class Job(BaseModel):
    job_title : str = Field(alias="job_title")
    experience_years : int = Field(alias="experience_years")
    education_level : str = Field(alias="education_level")
    skills_count : int = Field(alias="skills_count")
    industry : str = Field(alias="industry")
    company_size : str = Field(alias="company_size")
    location : str = Field(alias="location")
    remote_work : str = Field(alias="remote_work")
    certifications : int = Field(alias="certifications")
    salary : int = Field(alias="salary")
```
