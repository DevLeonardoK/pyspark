```python
from config import min_salary, max_salary
```

```python
from pydantic import BaseModel, Field, field_validator
```

```python
from enum import Enum
```

```python
class CompanySizes(str, Enum):
    startup = "Startup"
    small = "Small"
    medium = "Medium"
    large = "Large"
    enterprise = "Enterprise"
```

```python
class RemoteWorks(str, Enum):
    yes = "Yes"
    no = "No"
    hybrid = "Hybrid"
```

```python
class Job(BaseModel):
    job_title : str = Field(alias="job_title")
    experience_years : int = Field(alias="experience_years", gt=0, le=50)
    education_level : str = Field(alias="education_level")
    skills_count : int = Field(alias="skills_count", ge=0)
    industry : str = Field(alias="industry")
    company_size : CompanySizes = Field(alias="company_size")
    location : str = Field(alias="location")
    remote_work : RemoteWorks = Field(alias="remote_work")
    certifications : int = Field(alias="certifications", ge=0)
    salary : int = Field(alias="salary", ge=min_salary, le=max_salary)
    
    @field_validator("education_level",mode="before")
    def validation_education_level(cls, v):
        if v not in ("High School", "Diploma","Bachelor","Master","PhD"):
            raise ValueError("The input value is not valid")
    
```
